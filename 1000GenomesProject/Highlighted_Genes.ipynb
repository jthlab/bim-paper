```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('default')

from segmenter import segment_to_p
from browser import SPops, Pops

locvars = ['Chromosome', 'start', 'end']
```

```python
a = 0.05/311
key = 'btreeCmedi'

spopD = {'EAS': 'East Asian', 'EUR': 'European', 'AFR':'African', 'AMR': 'Ad-Mixed American', 'SAS':'South East Asian'}
```

```python
for spop in ['EAS', 'EUR', 'AFR', 'AMR', 'SAS']:
    dfs = []
    for i in range(26):
        if SPops[i] == spop:
            df = pd.read_csv('gene_scores/'+str(i)+'_'+key+'.csv')
            df.insert(0, 'Population', Pops[i][:3])
            dfs.append(df)

    dfs = pd.concat(dfs)   
    
    dx = dfs.sort_values('pbtreeCmedi').reset_index(drop = True).iloc[:100,:]

    cur_genes = []

    i = 0
    dels = []

    for i in range(dx.shape[0]):
        try:
            curg = dx.iloc[i]['genes'].replace(',','').split(' ')
        except:
            dels.append(i)
        if any([i in cur_genes for i in curg]):
            dels.append(i)

        cur_genes=cur_genes+curg

    dx = dx.drop(dels).iloc[:10,:].reset_index(drop=True)
    
    dx = dx.drop(columns = ['sbtreeCmedi'])
    
    dx['start'] = dx['start'].apply(lambda x: '{:0.2f}'.format(x/1e8))
    dx['end'] = dx['end'].apply(lambda x: '{:0.2f}'.format(x/1e8))
    
    dx['p'+key] = dx['p'+key].apply(lambda x: r'$'+'{:1.0}'.format(x).replace('e-', '\times10^{-')+'}$')
    dx['btreeCmedi'] = dx['btreeCmedi'].apply(lambda x: '{:0.2f}'.format(x))
    dz = dx.rename(columns = {'btreeCmedi':r'Average $\hat{\beta}$', 'pbtreeCmedi':'p value'})
    pd.set_option('max_colwidth', 400)
    print(dz.to_latex(index=False, escape = False, caption = 'Most significant directional selection hits for '+spopD[spop], column_format = 'lrllllp{6cm}'))  
```
